## Two-state Classical Systems and the Classical Bit

An abstract *two-state* classical system is any system that can be in one of *two* possible abstract states. Let $\{A, B\}$ be the set of possible abstract states. i.e. The system is either in state $A$ or in state $B$. An abstract two-state system can be implemented in different physical media as long as the physical medium can represent the two abstract states $A$ and $B$. To implement an abstract two-state system (with abstract states $A$ and $B$) in a physical medium, we use a property of the medium that can have two possible values. One value will represent the abstract state $A$ while the other value will represent abstract state $B$. The *bit*, the most basic unit of information in computing, is the amount of information you gain if the actual state of a two-state system is revealed to you (when the system has an equal chance of being in state $A$ or being in state $B$). It is the same amount of information you gain when you get an answer to a yes/no question when there is an equal chance of getting a 'yes' or getting a 'no' answer. The term 'bit' stands for "binary digit" or "binary information digit". A bit can either be the digit '0' or the digit '1'. When using the bit to represent the states of a two-state system, you use the '0' bit to represent one state and the '1' bit to represent the other state. It has been the convention to use the bit to represent states of a two-state system. ## Bit Strings

A *bit string* is simply a string of bits. e.g. $1$, $01$, $111$, $0001$. Bit string are used to encode information and the amount of information a bit string holds is the number of bits it has. The number of bits a bit string has is its *length*. i.e. The string $10010$ is bit string with 5 bits of information. It is a bit string of length $5$. Most digital data today like pictures, audio, video, and files are stored as bit strings (aka binary strings). In order to represent a bit string, you need multiple two-state systems with each system holding one bit of information. For example, a magnetic drive with a 1-gigabyte (1GB) capacity stores a bit string of length 8 billion (1 byte = 8 bits, 1 gigabyte = $1000^3$ bytes) so you need 8 billion two-state systems and for magnetic drives these two-state systems are magnetic drive cells where each cell is a 'magnetic domain' with a specific magnetic orientation that represent either '0' or '1'.

The set of symbols $\Sigma = \{0,1\}$ is called the *alphabet* of bit strings. The set $\Sigma^*$ (pronounced as "Sigma star") is the set of all bit strings (of any lenth). i.e. $\Sigma^* = \{\varepsilon, 0, 1, 00, 01, 10, 11, 000, 001, ...\}$. The $\varepsilon$ (or sometimes $\lambda$) symbol is used for the "empty string", the bit string of length 0. The set $\Sigma^+$ (pronounced as "Sigma plus") is the set of all bitstring excluding the empty string $\varepsilon$. The set $\Sigma^n$ is the set of bit strings of length $n$. i.e. $\Sigma^2 = \{00, 01,10, 11\}$. The *concatenation* operation is a binary operation that strings two bit strings together. i.e. Let $B_1$ and $B_2$ be bits strings, the the concatentation of $B_1$ with $B_2$, denoted by $B_1 | B_2$, is simply $B_1B_2$. e.g. If $B_1 = 111$ and $B_2 = 00100$, then $B_1 | B_ 2 = B_1B_2 = 11100100$.

## Two-State Quantum Systems and the Quantum Bit

In a two-state quantum system, similar to a two-state classical system, if you observe the system you will 'see' it in either one of two definite states.  Let $\{\ket{A}, \ket{B}\}$ be the set of observable abstract states. i.e. The system can be observed to be either in state $\ket{A}$ or in state $\ket{B}$.

An abstract two-state system can be implemented in different physical media as long as the physical medium can represent the two abstract states $A$ and $B$. To implement an abstract two-state system (with abstract states $A$ and $B$) in a physical medium, we use a property of the medium that can have two possible values. One value will represent the abstract state $A$ while the other value will represent abstract state $B$. Algebraic representation of the quantum state $\ket{\psi}$ using the basis $\{\ket{0}, \ket{1}\}$:

$$
\ket{\psi} = \alpha \ket{0} + \beta \ket{1} 
$$

In a two-level system with basis states $\{\ket{0}, \ket{1}\}$, by observing the system you will either see that system in state $\ket{0}$ or in state $\ket{1}$. The state
$\ket{\psi} = \alpha \ket{0} + \beta \ket{1}$ is not an actual state that you can observe but it gives you information about the probability of seeing the system in state
$\ket{0}$ and the probability of seeing the system in state $\ket{1}$ if you observe the system. The $\alpha$ in $\ket{\psi}$ is called the <i>amplitude</i> of the basis 
state $\ket{0}$ while the $\beta$ in $\ket{\psi}$ is amplitude of basis state $\ket{1}$. The amplitudes associated with the basis state tells you about the probability associated with seeing the particular basis state if observe the system. Getting the square of the amplitude of the a basis state (technically, getting the square modulus) will give you the probability of seeing that basis state if you observe the system. i.e. $|\alpha|^2$ is the probability of seeing the basis state $\ket{0}$ while $|\beta|^2$ is the probability of seeing the basis $\ket{1}$.

### Born Rule

The  <i>"Born Rule"</i> is a postulate in quantum mechanics that states the probability density of finding a system in a given state, when measured, is proportional to the square of the amplitude of the system's wavefunction at that particular state. If you have a state $\ket{\psi} = \alpha \ket{0} + \beta \ket{1}$

$$
\text{Born Rule:   }|\alpha|^2 + |\beta|^2 = 1
$$

## Multi-qubit States

### Example 1 ($Qubit_1 = 0$, $Qubit_2 = 0$):

State of $Qubit_1$: 
$$\ket{\psi} = \ket{0} = 1 \ket{0} + 0 \ket{1} = 
\begin{bmatrix} 
1 \\
0 
\end{bmatrix}
$$

State of $Qubit_2$: 
$$\ket{\phi} = \ket{0} = 1 \ket{0} + 0 \ket{1} = 
\begin{bmatrix} 
1 \\
0 
\end{bmatrix}
$$

Combined State of Qubit 1 and Qubit 2:




$$
\begin{align*} 

\ket{X} &= \ket{\psi} \otimes \ket{\phi}\\ 
        &= \ket{0} \otimes \ket {0} \\
        &= \ket{00} \\
        &=\begin{bmatrix} 
1 \\
0 
\end{bmatrix} \otimes
\begin{bmatrix} 
1 \\
0 
\end{bmatrix}\\ &= 
\begin{bmatrix} 
1 \begin{bmatrix} 
   1 \\
   0 
   \end{bmatrix}
\\
0 \begin{bmatrix} 
   1 \\
   0 
   \end{bmatrix}  
\end{bmatrix} = 
\begin{bmatrix} 
   1 \cdot 1 \\
   1 \cdot 0 \\
   0 \cdot 1 \\
   0 \cdot 0
   \end{bmatrix} =
\begin{bmatrix} 
   1 \\
   0 \\
   0 \\
   0 
   \end{bmatrix} 
\end{align*}$$
